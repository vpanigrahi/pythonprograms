```python
rc2={}
for i in range(len(r1)):
    if r1['class'][i]=='Hunter':
        p1=r1['date_time'][i]
        hour1=int(p1.split(' ')[1].split(':')[0])
        if hour1 in rc2.keys():
            rc2[hour1]+=1
        else:
            rc2[hour1]=1
        
        
    if i%50000==0:
        print(i*100.0/len(r1),'% done')
```

```python
# find max
for k,v in rc2.items():
    if v==max(rc2.values()):
        max1=k
        print k
```

# Answer 3

```python
rc3={}
rc3[1]=[]
Month_dict={}
PREV_MON=1
PREV_DATE=1
for i in range(len(r1)):
    
    p1=r1['date_time'][i]
    MON=int(p1.split(' ')[0].split('-')[1])
    if MON!=PREV_MON:
        Month_dict[PREV_MON]=rc3
        rc3={}
        PREV_MON=MON
        rc3[1]=[]
    DAT=int(p1.split(' ')[0].split('-')[2])
    if DAT!=PREV_DATE:
        rc3[DAT]=[]
        PREV_DATE=DAT
    avatar=r1['avatar_id'][i]
    if avatar not in rc3[DAT]:
        rc3[DAT].append(avatar)
    
        
        
    if i%50000==0:
        print(i*100.0/len(r1),'% done')
# finally for Dec
Month_dict[PREV_MON]=rc3
```

```python

for i in Month_dict.keys():
    for j in Month_dict[i]:
        Month_dict[i][j]=len(Month_dict[i][j])
        
```

```python
SE={}
Unique_class=list(set(r1['class']))
H_class={}
for i in range(len(Unique_class)):
    H_class[Unique_class[i]]=[]
    # print Unique_class[i]

for i in range(len(r1)):
    if r1['avatar_id'][i] not in SE.keys():
        SE[r1['avatar_id'][i]]={}
        SE[r1['avatar_id'][i]]['start']=r1['date_time'][i]
        SE[r1['avatar_id'][i]]['end']=r1['date_time'][i]
    else:
        SE[r1['avatar_id'][i]]['end']=r1['date_time'][i]
    
    H_class[r1['class'][i]].append(r1['avatar_id'][i])
    if i%50000==0:
        print(i*100.0/len(r1),'% done')
        
    
    
```

```python
from datetime import datetime
date_format = "%Y-%m-%d"
for i in range(len(SE.keys())):
    a = datetime.strptime(SE[SE.keys()[i]]['start'].split(' ')[0], date_format)
    b = datetime.strptime(SE[SE.keys()[i]]['end'].split(' ')[0], date_format)
    delta= b - a
    SE[SE.keys()[i]]['diff']=delta.days
```

```python
print 'annual average for all classes is',np.mean([SE[x]['diff'] for x in SE.keys()]),' days'
```

```python
for i in H_class.keys():
    H_class[i]=set(H_class[i])
#H_class
```

```python
Average_per_class={}
Variance_per_class={}
Actual_vals={}
for i in H_class.keys():
    p=[]
    for j in H_class[i]:
        p.append(SE[j]['diff'])
        
        #print p
    Actual_vals[i]=p
    Average_per_class[i]=np.mean(p)
    Variance_per_class[i]=np.var(p)
        
```
